```python
# 2. Import libraries
import pandas as pd
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
```

```python
# 3. Load your dataset (adjust the path if necessary)
df = pd.read_csv("/content/UpdatedResumeDataSet.csv", encoding='latin1')  # or try 'utf-8' or 'ISO-8859-1'
df.drop_duplicates(subset=['Resume', 'Category'], inplace=True)

df['Resume'] = df['Resume'].str.encode('latin1').str.decode('utf-8', 'ignore')
df.head()
```

```python
# 4. Vectorize the Resume column using TF-IDF
vectorizer = TfidfVectorizer(stop_words='english')
tfidf_matrix = vectorizer.fit_transform(df['Resume'])
```

```python
# 5. Define function to recommend job roles
def recommend_roles(user_input, df, vectorizer, tfidf_matrix, top_n=5):
    user_tfidf = vectorizer.transform([user_input])
    similarities = cosine_similarity(user_tfidf, tfidf_matrix).flatten()
    top_indices = similarities.argsort()[-top_n:][::-1]
    recommended = df.iloc[top_indices][['Category', 'Resume']].copy()
    recommended['Similarity'] = similarities[top_indices]
    return recommended
```

```python
# 6. Get user input and show recommendations
user_skills = input("Enter your skills (comma separated): ")
recommendations = recommend_roles(user_skills, df, vectorizer, tfidf_matrix)

print("\nTop Matching Job Roles:\n")
for i, row in recommendations.iterrows():
    print(f"Role: {row['Category']} (Score: {row['Similarity']:.2f})")
    print(f"Resume Sample: {row['Resume'][:200]}...\n")
```
